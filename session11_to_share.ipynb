# Long to wide


```python
import pandas as pd
```

```python


# Create a sample DataFrame for long data
df = pd.DataFrame({
    'id': [1, 1, 2, 2, 3, 3],
    'variable': ['A', 'B', 'A', 'C', 'A', 'B'],
    'value': [10, 20, 30, 40, 50, 60]
})

print(df)
# Convert the long data to wide data
df.pivot(index='id', columns='variable', values='value')
```

```python


# Create a sample DataFrame for long data
df = pd.DataFrame({
    'id': [1, 1, 2, 2, 3, 3],
    'variable': ['A', 'B', 'A', 'C', 'A', 'B'],
    'value': [10, 20, 30, 40, 50, 60],
    'value2': [101, 202, 330, 440, 650, 60]
})

print(df)
# Convert the long data to wide data
df.pivot(index='id', columns='variable', values=['value', 'value2'] )


```

# Wide to long

```python
# create data for wide data
df = pd.DataFrame({
    'group': [1, 2, 3],
    'A': [10, 30, 50],
    'B': [20, 40, 60],
    'C': [30, 50, 70]
})

print(df)
# Convert the wide data to long data
df.melt(id_vars='group', value_vars=['A', 'B', 'C'], var_name='variable', value_name='value')
```

```python
# create data for wide data
df = pd.DataFrame({
    'group': [1, 2, 3],
    'A': [10, 30, 50],
    'B': [20, 40, 60],
    'C': [30, 50, 70],
    'D': [301, 510, 720]
})

print(df)
# Convert the wide data to long data
df.melt(id_vars='group', value_vars=['A', 'B', 'C'], var_name='variable', value_name='value')
```

```python
from sklearn.datasets import load_wine
import pandas as pd

wine_data = load_wine()

print(wine_data.data.shape)
print(wine_data.target.shape)
print(wine_data.feature_names)
```

```python
# create a dataframe

# Creating a DataFrame for the features
df = pd.DataFrame(wine_data.data, columns=wine_data.feature_names)

# Adding the target column to the DataFrame
df['target'] = wine_data.target
```

```python
# read wine quality data set from web
df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-white.csv', sep=';')
df.head()
```

```python
df.columns = df.columns.str.replace(" ", "_")
```

```python
import statsmodels.formula.api as smf
from stargazer.stargazer import Stargazer

reg1 = smf.ols(formula="quality~alcohol", data=df).fit(cov_type="HC1")
reg2 = smf.ols(formula="quality~alcohol+residual_sugar", data=df).fit(cov_type="HC1")#
reg3 = smf.ols(formula="quality~alcohol+residual_sugar+pH", data=df).fit(cov_type="HC1")
stargazer = Stargazer([reg1, reg2, reg3])
stargazer.covariate_order(["alcohol", "residual_sugar", "pH", "Intercept"])
stargazer.rename_covariates({"alcohol": "Alcohol", "residual_sugar": "Residual Sugar", "pH": "pH Acidity Level", "Intercept": "Constant"})
stargazer
```

```python
# predict the quality of wine with linear regression
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import numpy as np

# split the data into train and test
train, test = train_test_split(df, test_size=0.2, random_state=123)

# fit a linear regression model
model = LinearRegression()
model.fit(train.drop('quality', axis=1), train['quality'])

# predict the quality of wine
pred = model.predict(test.drop('quality', axis=1))


test['predq']= pred

# calculate the RMSE
np.sqrt(mean_squared_error(test['quality'], pred))
```
