## Drop unnecessary columns

Delete the followings:

* everything with *tomato*
* Poster
* Website
* Response
* imdbID


```python
# Find columns that start with 'tomato'
columns_to_drop = [col for col in df.columns if col.startswith('tomato')]

# Drop these columns from the DataFrame
df = df.drop(columns=columns_to_drop)
df = df.drop(['Poster', 'Website', 'Response', 'imdbID'], axis=1, errors='ignore')
df.columns = [col.replace('.', '') for col in df.columns]
```

```python
df.columns
```

## Deal with column names

Rename some columns

* **Runtime** to **Lengt_min**

Make sure all names are lower case and are separeted by _ in case needed.


```python
df = df.rename(columns={'Unnamed: 0': 'IMDB_rank', 'Runtime': 'Lengt_min'})
```

## Create a month and day column. It should be int. The missing date should be january 1


```python
df['Released'] = df['Released'].fillna(pd.to_datetime('1900-01-01'))

df['Month'] = df['Released'].dt.month.astype(int)
df['Day'] = df['Released'].dt.day.astype(int)
```
